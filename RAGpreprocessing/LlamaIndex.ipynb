```python
from llama_index.core.node_parser import SentenceSplitter
from llama_index.core import Document
with open('output_ocr.txt', 'r', encoding='utf-8') as file:
    text = file.read()

# Wrap the text in a Document object
document = Document(text=text)

# Initialize the SentenceSplitter
parser = SentenceSplitter(chunk_size=250, chunk_overlap=20)

# Parse the document into nodes
nodes = parser.get_nodes_from_documents([document])

# Print the number of nodes created
print(f"Created {len(nodes)} nodes from 1 document")
```

```python
# Print the first few nodes
num_nodes_to_print = 5  # Adjust this number to print more or fewer nodes

for i, node in enumerate(nodes[:num_nodes_to_print]):
    print(f"Node {i + 1}:")
    print(node.get_text())  # Assuming get_text() is the method to retrieve the node's text
    print("-" * 40)  # Separator for readability
```
