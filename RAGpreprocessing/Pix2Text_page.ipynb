```python
import bs4, getpass, os, tiktoken
from langchain import hub
from langchain_chroma import Chroma
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import TextLoader
from langchain.prompts import ChatPromptTemplate

# Load, chunk and index the contents of the blog.
#loader = TextLoader("/content/output/output.md")
#docs = loader.load()

def num_tokens_from_string(string: str, encoding_name: str) -> int:
    """Returns the number of tokens in a text string."""
    encoding = tiktoken.get_encoding(encoding_name)
    num_tokens = len(encoding.encode(string))
    return num_tokens
#num_tokens_from_string(question, "cl100k_base")
def cosine_similarity(vec1, vec2):
    dot_product = np.dot(vec1, vec2)
    norm_vec1 = np.linalg.norm(vec1)
    norm_vec2 = np.linalg.norm(vec2)
    return dot_product / (norm_vec1 * norm_vec2)
#similarity = cosine_similarity(query_result, document_result)
```

```python
model_name = "BAAI/bge-small-en"
model_kwargs = {"device": "cuda:0"}
encode_kwargs = {"normalize_embeddings": True}
hf = HuggingFaceBgeEmbeddings(
    model_name=model_name, model_kwargs=model_kwargs, encode_kwargs=encode_kwargs
)
question = "What Ideal gas law?"
document = pages
```

```python
document
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
splits = text_splitter.split_documents(document)
vectorstore = Chroma.from_documents(documents=splits, embedding=hf)
retriever = vectorstore.as_retriever(search_kwargs={"k": 5})
```

```python
docs = retriever.get_relevant_documents(question)
len(docs)
```

```python
# Retrieve and generate using the relevant snippets of the blog.

# Prompt
template = """Answer the question based on the following context:
{context}

Question: {question}
"""

prompt = ChatPromptTemplate.from_template(template)


def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)

llm = HuggingFacePipeline.from_model_id(
    model_id="microsoft/Phi-3-mini-4k-instruct",
    task="text-generation",
    pipeline_kwargs={
        "max_new_tokens": 100,
        "top_k": 50,
        "temperature": 0.1,
    },
)
```

```python
#llm.invoke("Hugging Face is")

rag_chain = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)

rag_chain.invoke(question)
```

```python
import json
from pix2text import Pix2Text
from PIL import Image
import numpy as np

def element_to_json(element):
    """
    Convert an Element object to a JSON-compatible dictionary, handling special types.
    """
    return {
        'id': element.id,
        'box': element.box.tolist() if isinstance(element.box, np.ndarray) else element.box,  # Convert ndarray to list
        'type': element.type.name if hasattr(element.type, 'name') else element.type,  # Convert Enum to string
        'text': element.text,
        'isolated': element.isolated,
        'col_number': element.col_number,
        'score': element.score,
        'meta': element.meta if isinstance(element.meta, (dict, list, str, int, float)) else str(element.meta)  # Convert meta to string if it's complex
    }

def page_to_json(page):
    """
    Convert a Page object to a JSON-compatible dictionary.
    """
    return {
        'number': page.number,
        'id': page.id,
        'elements': [element_to_json(element) for element in page.elements]
    }

def document_to_json(doc):
    """
    Convert a Pix2Text Document object to a JSON-compatible dictionary.
    """
    return {
        'number': doc.number,
        'id': doc.id,
        'pages': [page_to_json(page) for page in doc.pages]
    }

# Convert the document to JSON
doc_json = document_to_json(doc)

# Save the JSON to a file
with open('output.json', 'w') as json_file:
    json.dump(doc_json, json_file, indent=4)

print("Document saved as 'output.json'")
```
